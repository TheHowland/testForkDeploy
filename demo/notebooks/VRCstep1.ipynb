Step response of R C circuit
============================

```python
from lcapy import Circuit
cct = Circuit()
cct.add('V1 1 0 u(t); down')
cct.add('R1 1 2 4; right')
cct.add('C1 2 3 0.1; down')
cct.add('W1 0 3; right')
```

```python
cct.draw()
```

```python
cct[2].V
```

```python
cct[2].v
```

```python
%matplotlib inline
cct[2].v.plot()
```

```python
cct[2].v.plot()
```

```python
cct.R1.i.plot()
```
