Step response of R L circuit
============================

```python
from lcapy import Circuit
cct = Circuit()
cct.add('V1 1 0 u(t); down')
cct.add('R1 1 2 4; right')
cct.add('L1 2 3 0.5; down')
cct.add('W1 0 3; right')
```

```python
cct.draw()
```

```python
cct.V[2]
```

```python
%matplotlib inline
cct.v[2].plot()
```

```python
cct.v[2].plot()
```

```python
cct.i['R1'].plot()
```
