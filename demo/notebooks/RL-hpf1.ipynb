```python
from lcapy import Circuit, j, omega
cct = Circuit()
cct.add("""
Vi 1 0_1; down
R 1 2; right, size=1.5
L 2 0; down
W 0_1 0; right
W 0 0_2; right, size=0.5
P1 2_2 0_2; down
W 2 2_2;right, size=0.5""")
```

```python
cct.draw()
```

```python
H = (cct.L.V / cct.Vi.V).simplify()
```

```python
H(j * omega)
```

```python
H(j * omega).magnitude
```

```python
H(j * omega).phase_degrees
```

```python
H1 = H.subs('L',1).subs('R',25e3)
H1(j * omega)
```

```python
from numpy import logspace
w = logspace(2, 6, 500)
%matplotlib inline
H1(j * omega).dB.plot(w, log_scale=True)
```

```python
H1(j * omega).phase_degrees.plot(w, log_scale=True)
```
