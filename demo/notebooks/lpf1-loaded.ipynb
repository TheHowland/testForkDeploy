```python
from lcapy import Circuit, j, omega
cct = Circuit()
cct.add("""
V1 1 0_1; down
Rs 1 2; right, size=1.5
C 2 0; down, v=v_C
W1 0_1 0; right
W2 0 0_2; right
P1 2_2 0_2; down
RL 2_2 0_2; down
W3 2 2_2;right""")
cct.draw()
```

```python
H = cct.V['RL'] / cct.V['V1']
```

```python
H(j * omega)
```

```python

H(j * omega).magnitude
```

```python
H(j * omega).phase
```

```python
H1 = H.subs('C', 1e-9).subs('Rs', 25e3).subs('RL', 5e3)
H1(j * omega)
```

```python
from numpy import logspace
w = logspace(2, 6, 500)
%matplotlib inline
H1(j * omega).dB.plot(w, log_scale=True)
```

```python

H1(j * omega).phase.plot(w, log_scale=True)
```
