```python
from lcapy import Circuit, j, omega
cct = Circuit()
cct.add("""
V1 1 0_1; down
Rs 1 2; right, size=1.5
C 2 0; down, v=v_C
W1 0_1 0; right
W2 0 0_2; right
Rin 2_2 0_2; down, v=v_{in}
W3 2 2_2; right
E1 3 0_3 2 0 A; down, l=A v_{in}
Rout 3 4; right, size=1.5
RL 4 0_4; down
W4 0_2 0_3; size=1.2
W5 0_3 0_4
P1 4 0_4; down
""")
cct.draw()

```

```python
H = cct.V['RL'] / cct.V['V1']
H(j * omega)
```

```python
H(j * omega).magnitude
```

```python
H(j * omega).phase
```

```python

H1 = H.subs({'C': 1e-9, 'Rs': 25e3, 'RL': 5e3, 'Rin' : 500e3, 'Rout' : 1e3, 'A' : 2})
H1(j * omega)
```

```python
from numpy import logspace
w = logspace(2, 6, 500)
%matplotlib inline
H1(j * omega).dB.plot(w, log_scale=True)
```

```python
H1(j * omega).phase.plot(w, log_scale=True)
```
