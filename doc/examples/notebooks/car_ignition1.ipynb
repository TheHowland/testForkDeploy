```python
from lcapy import Circuit
cct = Circuit()
cct.add("""
V 1 0 10; down
C 1 2 1e-6 0; right, size=2
R 2 3 2; down, size=1.5
L 3 0_1 8e-3 5; down, v=v_L, i=i
W 0 0_1; right
""")
cct.draw()
```

```python
scct = cct.s_model()
scct.draw()
```

```python
cct.L.V.s.partfrac()
```

```python
cct.L.v
```

```python
cct.L.I.s.poles()
```

```python
%matplotlib inline
from numpy import linspace
t = linspace(0, 10e-3, 400)
ax = cct.L.v.plot(t)
```

Note, in practice, there is a 100:1 step up to a secondary coil conencted to a spark-gap.  Once the spark forms, the secondary coil is shorted out and energy is removed from the L-Circuit damping the oscillation.
