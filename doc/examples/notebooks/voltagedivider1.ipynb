Resistive voltage divider
=========================

```python
from lcapy import Circuit
cct = Circuit()
cct.add('V1 1 0 {u(t)}; down')
cct.add('R1 1 2 4; right=1.5')
cct.add('R2 2 3 6; down=1.5')
cct.add('W1 0 3; right')
```

```python
cct.draw()
```

```python
cct[2].v
```

```python
%matplotlib inline
ax = cct[2].v.plot()
```

```python
ax = cct.R1.i.plot()
```
