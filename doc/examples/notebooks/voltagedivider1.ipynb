Resistive voltage divider
=========================

```python
from lcapy import Circuit
cct = Circuit("""
V1 1 0 {u(t)}; down
R1 1 2 4; right=2
R2 2 3 6; down=2
W1 0 3; right
""")
cct.draw()
```

```python
cct[2].v
```

```python
%matplotlib inline
ax = cct[2].v.plot()
```
