Step response of R L circuit
============================

```python
from lcapy import Circuit, s
cct = Circuit()
cct.add('V1 1 0 {u(t)}; down')
cct.add('R1 1 2 4; right=1.5')
cct.add('L1 2 3 0.5; down=1.5')
cct.add('W1 0 3; right')
```

```python
cct.draw()
```

```python
cct[2].V(s)
```

```python
%matplotlib inline
ax = cct[2].v.plot()
```

```python
ax = cct[2].v.plot()
```

```python
ax = cct.R1.i.plot()
```
